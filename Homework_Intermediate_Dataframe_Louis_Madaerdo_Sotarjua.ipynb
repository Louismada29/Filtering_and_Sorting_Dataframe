# DataFrame Manipulation

```python
datapass = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
datapass
```

## Filter and Sort DataFrame

```python
# make a filtering of dataframe to only show columns 'Name','Sex','Age','Ticket','Fare'
# then sort the dataframe by Age(oldest to youngest) and Fare (expensive to inexpensive)
datapass.filter(items=['Name','Sex','Age','Ticket','Fare']).sort_values(by=['Age','Fare'], ascending = False)
```
